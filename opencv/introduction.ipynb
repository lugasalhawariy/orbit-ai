```python
# konversi gambar dari BGR ke RGB
img2 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(8,8))
# warna sudah dirubah ke RGB dengan fungsi cv.COLOR_BGR2RGB
plt.imshow(img2)
```

```python
# Rotation
center = img2.shape[1]/2, img2.shape[0]/2
angle = 30
scale = 1.0

M = cv.getRotationMatrix2D(center, angle, scale)

rotate_cat = cv.warpAffine(img2, M, (int(img2.shape[1]*1.3), int(img2.shape[0]*1.3)), borderValue=(255,255,255))
plt.imshow(img2)
plt.figure()
plt.imshow(rotate_cat)
```
