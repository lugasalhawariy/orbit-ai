```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# menghitung situs (rumus)
def sinusGenerator(amplitudo, frekuensi, waktu, theta):
    t = np.arange(0, waktu, 0.1)
    y = amplitudo * np.sin(2 * frekuensi * t + np.deg2rad(theta))
    return t,y
```

```python
# membuat plot
t1, y1 = sinusGenerator(1,1,4,0)
t2, y2 = sinusGenerator(1,1,4,30)
t3, y3 = sinusGenerator(1,1,4,60)
```

```python
data1 = plt.plot(t1,y1)
plt.setp(data1, color='r', linestyle='--', linewidth='1.75')
data2 = plt.plot(t2,y2)
plt.setp(data2, color='c', linestyle='-', linewidth='2.75')
data3 = plt.plot(t3,y3)
plt.setp(data3, color='p', linestyle='-', linewidth='3.75')
# setting properties


```
